# Bytes
1. bytes data type represents a group of bytes numbers just like a array.
2. The only allowed values for bytes data types are 0 to 255.
3. Once we create bytes data types we cannot change its value. 

```python
l = [10,20,30,40,50]
b = bytes(l)
print(type(b))

print("Elements in b")
for x in b:
    print(x)

```

```python
print(b[0])

print("Slicnig is possible")
b1 = b[1:3]

for x in b1:
    print(x)
```

```python
print("Immutable Data types")
b[1] = 50
```

# bytearray
1. bytearray is excalty same as bytes data type expect that its element can be modified 

```python
l = [10,20,30,40,50,60]
b = bytearray(l)
print("Elements for Bytearray")
for x in b:
    print(x)
```

```python
b[3] = 250

print()
```
