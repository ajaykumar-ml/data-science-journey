# Set
1. If we want to represent a group of  values without duplicates where order is not important then we should go for set Data type
2. Insertion order is not preserved
3. Heterogeneous objects are allowed
4. It is mutuable collection
5. Growable in nature

example: s1 = {1,2,3} , s2 = {2,3,1}, s3 = {2,1,3} where s1,s2 and s3 are same set.


```python
# empty set
s = set()
print(type(s))

s = {} #empty dictionary
print(type(s))
```

```python
s = {1,2,3,4,5,"ajay",1,2}
print(type(s))
print(s)
```

**“No order → no index → no indexing.”**

```python
s = {1,2,3,4,5,"ajay",1,2}
print(s[1])
```

```python
#empty set
s = set()
s.add(1)
s.add(2)
s.add(3)
s.add("ajay")
print(s)
```

```python
s = set()
s.add(1)
s.add(2)
s.add(3)
s.add("ajay")
print(s)

s.remove("ajay")
print(s)
```

### add() vs append()

| Feature        | append()   | add()          |
| -------------- | ---------- | -------------- |
| Data structure | List       | Set            |
| Add position   | End        | Anywhere       |
| Order          | Maintained | Not maintained |
| Duplicate      | Allowed    | Not allowed    |
| Multiple items | ❌          | ❌              |


# frozenset
1. It is exaclty same as set except that it is immutable
2. Hence we cannot use add or remove functions

```python
s = {10,"ajay",20,50,"kumar",10}
fs = frozenset(s)
print(type(fs))
print(fs)
```
