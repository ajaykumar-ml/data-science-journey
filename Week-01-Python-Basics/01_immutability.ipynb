# Immutability

 id(): यह किसी object की identity लौटाता है, यानी memory में उस object का unique address (reference).<br>
 id(): is used to know whether two variables point to the same object in memory.

```python
x = 10
print(id(x))
x = x+1
print("New Object will be created")
print(id(x))
```

```python
x = 10
y = x
print(id(x))
print(id(y))
y = y+1
print("With this change new object will created")
print(id(x))
print(id(y))
```

```python
x = 10
y = 10
print(x  is y)
print("x and y are pointing to same object")
```

```python
a = "ajay"
b = "ajay"

print(a is b)

a = "ajay kumar"
b = "ajay kumar"

print(a is b)

print("Interviewer keyword सुनना चाहता है: String interning is implementation dependent.")
```

**Manual Interning**

```python

import sys

a = sys.intern("ajay kumar")
b = sys.intern("ajay kumar")

print(a is b)
```

**NOTE:**
Interning is a memory optimization where Python reuses immutable objects, but is should never be used for value comparison.

```python
a = 10+20j
b = 10+20j
print(a is b)
```

**NOTE:** In all fundamental data types complex is not immutable.
Fundamental data types are int, string , bool, float.

**Example of Mutable object.**

```python
l1 = [10,20,30,40]

print("ID :",id(l1))
l1[2] = 50
print(l1)
print("ID :",id(l1))
print("List is Mutable")
```

```python
l1 = [10,20,30,40]
l2 = l1
print("ID :",id(l1))
print("ID :",id(l2))

print("\nBoth l1 and l2 refrencing to same list object ")

print(l1)
print(l2)

l1[1] = 50
print("\n")

print("ID :",id(l1))
print("ID :",id(l2))

print("\nBoth l1 and l2 refrencing to same list object")

print(l1)
print(l2)
```
