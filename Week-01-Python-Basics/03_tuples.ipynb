# Tuple 

1. tuple data type is same as list data type except that is immmutable. i.e **we cannot change values.**
2. tuple elements can be represented within parenthesis.
3. tuple is read only version of list.
   

```python
#empty tuple
t = ()

print(type(t))
```

```python
t = (10,"ajay",20,30)

print(t)

print(t[1])
print(t[1:3])
```

```python
t = (10,"ajay",20,30)

t[1] = 100

#TypeError: 'tuple' object does not support item assignment
```

```python
t.append(50)
```

```python
t.remove()
```

### ⚠️ Single element tuple (important)

t = (10,)   # comma जरूरी है
print(type(t))

1. ❌ (10) tuple नहीं, int है
2. ✅ (10,) सही tuple है

```python
t = (10)

print(type(t))

t = (10,)
print(type(t))
```
