### Relational Operators
1. In Python, relational operators (also called comparison operators) are used to compare two values.
2. | Operator | Meaning                  | Example  |
| -------- | ------------------------ | -------- |
| `==`     | Equal to                 | `a == b` |
| `!=`     | Not equal to             | `a != b` |
| `>`      | Greater than             | `a > b`  |
| `<`      | Less than                | `a < b`  |
| `>=`     | Greater than or equal to | `a >= b` |
| `<=`     | Less than or equal to    | `a <= b` |


```python
a = 10
b = 20

print(a == b)   # False
print(a != b)   # True
print(a > b)    # False
print(a < b)    # True
print(a >= 10)  # True
print(b <= 15)  # False
```

### In Python, strings are compared using relational operators, and the comparison is done lexicographically (character by character, based on Unicode/ASCII values).

How Python Compares Strings

Python compares:

1. First character of both strings

2. If equal ‚Üí move to next character

3. Stops when a difference is found

```python
a = "apple"
b = "banana"

print(a == b)   # False
print(a != b)   # True
print(a < b)    # True
print(a > b)    # False
```

```python
print("car" > "cat")   # ('t' > 'r')
```

### Equality vs Identity (== vs is) Important 

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True  (same values) (check content)
print(a is b)  # False (different objects)
print(id(a))
print(id(b))

print("")

c = 10
d = 10
print(c == d)
print(c is d)
print(id(c))
print(id(d))
```

### Logical Operators in Python
| Operator | Meaning                                    |
| -------- | ------------------------------------------ |
| `and`    | True if **both** conditions are True       |
| `or`     | True if **at least one** condition is True |
| `not`    | Reverses the result                        |



```python
a = 10
b = 20

print(a > 5 and b > 15)   # True
print(a > 5 and b > 25)   # False
```

```python
a = 10
b = 5

print(a > 5 or b > 10)    # True
print(a < 5 or b > 10)    # False
```

```python
x = True

print(not x)   # False
```

### Logical Operators with Non-Boolean Values

Python uses truthy and falsy values.
Falsy values:

1. False
2. 0
3. None
4. ""
5. [], {}, ()

***How Logical Operators Work***
1. **and** Operator (x and y)</br>
   .If x evaluates to false , then result is x</br>
   .If x evaluates to true , then result is y</br>

   OR

   .Returns the first falsy value</br>
   .If all values are truthy ‚Üí returns the last value
   




```python
print(10 and 20)      # 20
print(0 and 5)        # 0
print("hi" and "")    # ""
```

2. **or** Operator</br>
   .If x evaluates to true then result is x otherwise result is y.</br>

   OR</br>
   
   .Returns the first truthy value</br>
   .If all values are falsy ‚Üí returns the last value

```python
print(0 or 5)         # 5
print("" or "hello")  # hello
print(None or 0)      # 0
```

3. **not** Operator <br>
   .Always returns a boolean<br>
   .Reverses truth value

```python
print(not 0)        # True
print(not "text")   # False
print(not [])
print(not "[1]")
print(not "[]")
```
