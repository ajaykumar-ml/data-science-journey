# Range
1. range data types represents a sequence of numbers
2. The elements presents in range data types are not modifiable, i.e range data types is immutable

```python
# Form 1: range(10)

r = range(10)
print(type(r))

for x in r:
    print(x)
```

```python
# Form 2: range(begin , end)
r = range(1,10)
print(r)
print("\n Elements in range:\n")
for x in r:
    print(x)
```

```python
# Form 3: range(begin, end, increment/decrement)

print("Increment Example")
r = range(1,21,5)
for x in r:
    print(x)
print("Decrement Example")
r = range(21,1,-4)
for x in r:
    print(x)
```

**Indexing and Slicing is possible**

```python
r = range(10,31)
print(r[2])
print(r[-1])
print(r[1:5])
```

```python
r[1] = 11
```
