### 3Ô∏è‚É£ if‚Äìelif‚Äìelse Statement 
üìå Python checks conditions top to bottom<br>
üìå First True condition executes

````python
per = float(input("Enter your percentile"))
if per >=99:
    print("Eligible for JEE Advance")
elif  90<= per <99:
    print("Eligible for JEE Mains Counselling")
else:
    print("Better luck next time")
````

### 4Ô∏è‚É£ Nested if Statement
. An if statement inside another if.

````python
num  = int(input("Enter an integer"))
if num>0:
    if num%2==0:
        print("Positive even integer")
    else:
        print("Positive odd integer")
else:
    print("Negative Integer")
````

### Example  Mean (Average) of a List of Marks:

````python
marks = [88,97,77,90,36,99]
length = 0
total = 0 

for x in marks:
    length += 1
    total += x

avg = total/length

print("Mean:", avg)
````

***PROBLEM STATEMENT***

Find the sum of all even natural numbers between 1 and n (inclusive), where n is a positive integer given by the user.

````python
num = int(input("Enter a positive integer"))

total = 0

for x in range(1,num+1):
    if x%2 == 0:
        total += x

print("Sum of all even natural number is:", total)

````

***Problem Statement***

Given a positive integer n, find the sum of all numbers between 1 and n (inclusive) that are divisible by 3 or 5.

````python
num = int(input("Enter a positive integer: "))
total = 0

for x in range(1, num + 1):
    if x % 3 == 0 or x % 5 == 0:
        total += x

print("Sum is:", total)
````

***(inclusion‚Äìexclusion principle)***

````python
num = int(input("Enter a positive integer: "))
three_total = 0
five_total = 0
fifteen_total = 0

for x in range(1, num+1):
    if x%3 ==0:
        three_total += x
    if x%5 == 0:
        five_total += x
    if x%15 ==0:
        fifteen_total += x
        
# Inclusion‚ÄìExclusion Principle        
total = three_total + five_total - fifteen_total
print("Sum is:", total)
````

This problem is solved using two approaches:

1. Mathematical formula (Gauss method)
2. Iterative approach using a while loop


````python
num = int(input("Enter a positive integer"))

total = 0

while(num>=1):
    total += num
    num -= 1
    
print("Sum is:", total)    
````

### Nested for Loop

***Problem Statement*** Given a positive integer n, print the multiplication tables from 1 to n using a nested for loop.

````python
num = int(input("Enter a positive integer"))

for i in range(1,num+1):
    print("Table of ", i)
    for j in range(1,11):
        print(i, "x", j, "=", i*j)
````

***Problem Statement: Square Star Pattern***

Given a positive integer n, write a program to print a square pattern of stars (*) of size n √ó n using nested loops.

````python
num = int(input("Enter a positive number"))

for i in range(1,num+1):
    for j in range(1,num+1):
        print("*",end=" ")
    print()
````

### Problem  Using break
***First Multiple of 7 Greater Than 50***

Starting from 1, find the first number greater than 50 that is divisible by 7.
Stop the search as soon as the number is found.

````python
num = 51
while True:
    if num > 51 and num%7 == 0:
        break
    num+=1
print("Required number is :",num)    
````

### for‚Äìelse in Python
1. The else block executes only if the for loop finishes normally (without break)

Python does this:
1. Start the loop
2. Run each iteration
3. If break occurs ‚Üí skip else
4. If loop ends naturally ‚Üí execute else

***Smallest Divisor of a Number (Except 1)***

Given a positive integer n > 1, find the smallest divisor of n greater than 1.
Stop checking further once the divisor is found.

````python
num = int(input("Enter a positive number greater than 1"))

for x in range(2,num+1):
    if(num%x == 0):
        print("Smallest divisor of {} is {}".format(num,x))
        break
else:
    print(f"{num} is Prime Number")
````

### Problem Based on continue
***Print All Odd Numbers Between 1 and n***

Given a positive integer n, print all odd numbers from 1 to n, skipping even numbers.

````python
num = int(input("Enter a positive integer"))
for x in range(1,num+1):
    if x%2 == 0:
        continue
    print(x)
````

***Sum of Numbers Excluding Multiples of 3***

Given a positive integer n, find the sum of all numbers from 1 to n, excluding numbers that are divisible by 3.

````python
num = int(input("Enter a postive number"))
total = 0
for x in range(1,num+1):
    if x%3 == 0:
        continue
    total += x

print("Sum is {}".format(total))    
````
