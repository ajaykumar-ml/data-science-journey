| Data Type     | Description                                   | Mutable | Insertion Order Preserved | Indexing | Slicing | Duplicates |

| int          | Integer numbers (whole numbers)               | ❌ No   | ❌ No                    | ❌ No   | ❌ No  | ❌ No     |
| float        | Decimal / floating point numbers              | ❌ No   | ❌ No                    | ❌ No   | ❌ No  | ❌ No     |
| complex      | Complex numbers (a + bj)                       | ❌ No   | ❌ No                    | ❌ No   | ❌ No  | ❌ No     |
| bool         | Boolean values (True / False)                 | ❌ No   | ❌ No                    | ❌ No   | ❌ No  | ❌ No     |
| str          | Sequence of characters                        | ❌ No   | ✅ Yes                   | ✅ Yes  | ✅ Yes | ✅ Yes    |
| list         | Ordered, mutable collection                   | ✅ Yes  | ✅ Yes                   | ✅ Yes  | ✅ Yes | ✅ Yes    |
| tuple        | Ordered, immutable collection                 | ❌ No   | ✅ Yes                   | ✅ Yes  | ✅ Yes | ✅ Yes    |
| set          | Unordered collection of unique elements       | ✅ Yes  | ❌ No                    | ❌ No   | ❌ No  | ❌ No     |

| bytes        | Immutable sequence of bytes                   | ❌ No   | ✅ Yes                   | ✅ Yes  | ✅ Yes | ✅ Yes    |
| bytearray    | Mutable sequence of bytes                     | ✅ Yes  | ✅ Yes                   | ✅ Yes  | ✅ Yes | ✅ Yes    |

2. “जो लाइन में हैं वही गिनती में हैं”

Indexing ✅ → str, list, tuple, bytes, bytearray

# None

What is None?

None एक special built-in object है

इसका type होता है NoneType

यह 0, empty string, या False नहीं है

```python
x = None
print(id(x))
print(type(x))
```

1. Why None is used?

✔️ Default value set करने के लिए
✔️ Function failure / no result बताने के लिए
✔️ Optional parameters के लिए
