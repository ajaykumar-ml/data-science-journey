### Input function
1. In Python 2 the following 2 function are available to read dynamic input from the keyboard <br>
a). raw_input() <br>
b). input()

2. **raw_input():** This fucntion reads the data from the keyboard in the form of String format. We have to convert that string type to our required type by using the corresponding type casting methods.

3. **input():** input function can be used to read the data directly in our required format. We are not required to perform type casting.

4. NOTE: But in python 3 we have only input() method and raw_input() method is not available. Python 3 input() method behaviour exactly same as raw_input() method of python 2 i.e every input is treated as str type only.

5. **Comparison Table**
| Feature        | `raw_input()` | `input()` (Py2) | `input()` (Py3) |
| -------------- | ------------- | --------------- | --------------- |
| Python Version | Python 2      | Python 2        | Python 3        |
| Returns        | `str`         | Evaluated value | `str`           |
| Safe           | ‚úÖ Yes        | ‚ùå No            | ‚úÖ Yes          |
| Exists Today   | ‚ùå Removed    | ‚ùå Removed       | ‚úÖ Used         |

### üîπ Taking Multiple Values from Keyboard in Python

### How Python Reads This (Order of Execution)

input() ‚Üí "10 20"

.split() ‚Üí ["10", "20"]

List comprehension ‚Üí [10, 20]

Unpacking ‚Üí a = 10, b = 20

```python
nums = input("Enter two numbers") # Takes input from keyboard , Always returns a string

print(nums)

l1 = nums.split() #Splits string by spaces , Returns a list of strings

print(l1)

l2 = [int(x) for x in l1] # Take each element x , Convert it to int, Store in a new list

print(l2)

a,b  = l2 # a gets 10, b gets 20

print("a=", a,"b=", b, "Sum=",a+b)
```

```python
# above lines can be written in single line
a,b = [int(x) for x in input("Enter two numbers").split()] 
print("a=", a,"b=", b, "Sum=",a+b)
```

### ‚ö†Ô∏è Possible Errors (VERY IMPORTANT)

```python
a, b = [int(x) for x in input().split()]

```

```python
#  Alternative (Using map())

a, b = map(int, input("Enter two numbers: ").split())
```
