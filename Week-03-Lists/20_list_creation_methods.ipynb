1️⃣ **Creating a List Using Square Brackets []**

```python
numbers = [1, 2, 3, 4, 5]
names = ["Ajay", "Rahul", "Priya"]
mixed = [1, "Python", 3.14, True]
numbers, names, mixed
```

2️⃣ **Creating a List Using the list() Constructor**
1. The list() function converts an iterable into a list.
2. Used to convert other data types into lists.

```python
list_from_tuple = list((10,20,30))

# From string
list_from_string = list("Python")

list_from_tuple, list_from_string
```

3️⃣ **Creating a List Using range()**
1.  range() generates a sequence of numbers, which can be converted into a list.

```python
numbers = list(range(0,14))
even_numbers = list(range(0,14,2))
numbers,even_numbers
```

4️⃣ **Creating a List Using List Comprehension**

```python
numbers = [int(x) for x in input("Enter integers").split()]
numbers
```

```python
## Summary
| Method             | Example                           |
| ------------------ | --------------------------------- |
| Square brackets    | `[1, 2, 3]`                       |
| `list()`           | `list("abc")`                     |
| `range()`          | `list(range(5))`                  |
| List Comprehension | `[x*x for x in range(5)]`         |
| User Input         | `list(map(int, input().split()))` |

```
