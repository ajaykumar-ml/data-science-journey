# NUMERIC OPERATIONS ON LISTS
- Numeric operations work only on lists containing numeric values (int, float).

### sum() — Total of All Elements
1. Adds all numeric elements of a list
2. Returns a single numeric value
3. Does not modify the list

```python
expenses = [1232,456, 654,764]
sum(expenses)
```

### min() — Smallest Value
1. Returns the smallest element in the list
2. List must contain comparable values

```python
temperatures = [32, 28, 35, 30]
min(temperatures)
```

### max() — Largest Value
1. Returns the largest element in the list

```python
scores = [65, 88, 72, 91]
max(scores)
```

# MEMBERSHIP OPERATIONS IN LISTS
1. Membership operations check presence or absence of elements.

### in — Check if Element Exists
1. Returns True if element exists in the list
2. Returns False otherwise

### not in — Check if Element Does NOT Exist

1. Returns True if element is absent
2. Returns False if present

```python
subjects = ["Mathematics", "Chemistry", "Physics"]
"Mathematics" in subjects
```

```python
"Biology" not in subjects
```

## Summary
| Category   | Operation | Purpose           |
| ---------- | --------- | ----------------- |

| Numeric    | `min()`   | Smallest value    |
| Numeric    | `max()`   | Largest value     |

| Membership | `not in`  | Check absence     |
