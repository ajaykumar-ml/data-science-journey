#### What is List Comprehension? 
1. List comprehension is a short and readable way to create a new list from an existing iterable.
2. **Basic Syntax** : [new_value **for** item **in** iterable]
3.
| Part        | Meaning                |
| ----------- | ---------------------- |
| `new_value` | What you want to store |
| `item`      | Each element           |
| `iterable`  | List, range, etc       |


```python
marks = [78, 29, 92, 31, 74, 18, 88, 25, 69, 76]
```

```python
updated_marks = [x+5 if 28<=x<33 else x for x in marks]
updated_marks
```

### enumerate() in List Comprehension

```python
names = [ "Rahul Sharma", "Ankit Verma","Ajay Kumar","Rohit ", "Sandeep ","Neha ", "Pooja Mehta", "Priya Patel", "Anjali Jain", "Kavita " ]
marks = [78, 29, 92, 31, 74, 18, 88, 25, 69, 76]
indexed_names = [f"{index+1}. {name}" for index, name in enumerate(names)]
indexed_names 
```

### Using zip() in List Comprehension

```python
for name, mark in zip(names,marks):
    print(f"{name:<15} {mark:>3}")
```
