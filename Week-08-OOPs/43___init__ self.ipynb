# What is `__init__` Method?

1. __init__ is a special method in Python used to initialize an object.
2. It is automatically called when an object is created
3. It helps us assign initial values to object variables
4. It is also called a constructor

**Basic Syntax:**
``` Python
    class ClassName:
        def __init__(self, parameters):
            self.variable = value

**Example 1: Student Class (Basic)**

````python
class Student:
    def __init__(self,name,rollno,course):
        self.name = name
        self.rollno = rollno
        self.course = course

    def showData(self):
        print("Name:", self.name, "Roll No:" ,self.rollno, "Course:", self.course)
````

````python
student1 = Student("ajay",101,"DSML")
student2 = Student("anushaka",102,"AI")
````

````python
student1.showData()
student2.showData()
````

## Self Variable in Python
- self represents the current object.
- `__init__` runs automatically
- Attributes are created per object
- self connects variables to the object
- Each object has its own data

#### Important Notes About self

1️⃣ **self must be the first parameter of the constructor**

````python
def __init__(self):
    pass
````

2️⃣ **self must be the first parameter of all instance methods**

````python
def showData(self):
    pass
````

## What is a Constructor?

A constructor is a special method that is executed automatically when an object is created.

In Python, the constructor is named:
`__init__`

**NOTE:**
- It is executed at the time of object creation
- Used to declare and initialize instance variables
- Runs once per object
- Can take any number of arguments (minimum one: self)
- If not defined, Python provides a default constructor

### Relationship Between self and Constructor

- self → refers to current object
- Constructor → initializes data inside that object
