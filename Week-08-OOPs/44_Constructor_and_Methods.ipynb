# Definition: Constructor
A constructor is a special method in a class that is automatically executed when an object is created. It is mainly used to initialize
(instance) variables of the object.

- In Python, the constructor is named `__init__`
- It runs only once per object
- It prepares the object for use
- When you create an object, Python:
    - Allocates memory for the object
    - Automatically calls `__init__`
    - Stores initial data inside the object

# Definition: Method
A method is a function defined inside a class that represents the behavior or actions an object can perform.
- Methods are called explicitly using an object
- They can be called multiple times
- They work on data already stored in the object

```python
class Dataset:
    def __init__(self, name, rows, column):
        self.name = name
        self.rows = rows
        self.column = column
    
    def show_info(self):
        print("Dataset:", self.name)
        print("Rows:", self.rows)
        print("Columns:", self.column)

    def is_large_dataset(self):
        return self.rows > 10000       
```

```python
sales_dataset = Dataset("Sales 2025", 200000,500)
sales_dataset.show_info()
sales_dataset.is_large_dataset()
```
