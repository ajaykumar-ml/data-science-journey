# Different Types of Methods in Python 
- **Python supports three main types of methods:**
    1. Instance Method
    2. Class Method
    3. Static Method

### 1Ô∏è‚É£ Instance Method
An instance method is a method that works on instance variables and is called using an object of the class.<br>
**üîπ Explanation**
- First parameter is always self
- Can access:
    - Instance variables
    - Class variables
- Represents object-specific behavior

**Example:**

```python
class Dataset:
    def __init__(self,name,rows):
        self.name = name
        self.rows = rows
    
    def show_dataset(self):
        print("Dataset:", self.name)
        print("Rows:", self.rows)

    def is_large(self):
        return self.rows >= 10000

dataset1 = Dataset("Sales", 150000)
dataset2 = Dataset("Students", 1000)

dataset1.show_dataset()
dataset2.show_dataset()
```

```python
dataset1.is_large(), dataset2.is_large()
```

**Why Instance Method here?** <br> Because dataset size is different for each dataset object.

### 2Ô∏è‚É£ Class Method
A class method is a method that works on class variables and is called using the class name.<br>

üîπ **Explanation**
- Uses `@classmethod` decorator
- First parameter is cls (class reference)
- Can access:
    - Class variables
- Used for class-level operations

**Example**

```python
class MLModel:
    algorithm = "Linear Regression"

    @classmethod
    def change_algorithm(cls,new_algorithm):
        cls.algorithm = new_algorithm

MLModel.change_algorithm("Decision Tree")
MLModel.algorithm
```

**Why Class Method here?** <br>
Because algorithm type is shared by all model objects.

### 3Ô∏è‚É£ Static Method
A static method is a method that does not depend on instance variables or class variables. <br>
**üîπExplanation**
- Uses `@staticmethod` decorator
- Does NOT take self or cls
- Behaves like a normal function
- Kept inside a class for logical grouping

**Example**

```python
class DataUtils:

    @staticmethod
    def is_valid_percentage(value):
        return 0 <= value <= 100

print(DataUtils.is_valid_percentage(85))
print(DataUtils.is_valid_percentage(150))
```

**üîπ Can We Call Static Method Using Object?**

```python
obj = DataUtils()
print(obj.is_valid_percentage(90))
```

‚úî This works <br>
‚ùå But should be avoided

**üîπ Why Calling via Class Name Is Better ?** <br>
Calling a static method using the class name makes it clear that:
- The method does not depend on object data
- The method does not depend on class data
- It is just a utility function logically grouped inside the class


**NOTE:** Yes, static methods are ideally called using the class name,
because they do not depend on instance or class state.
