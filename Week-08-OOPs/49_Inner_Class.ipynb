# Inner Class in Python
- An inner class is a class defined inside another class. It is mainly used to logically group classes that are closely related.
- Syntax
```python
    class OuterClass:
    
        class InnerClass:
            pass


### üè¶ Real-Life Example: Bank and Account
1. **Think about this:**
    - A Bank has multiple Accounts
    - An Account cannot exist independently without a Bank
    - So logically, Account belongs inside Bank
2.  **Why Inner Class Here?**
    - Account is tightly related to Bank
    - It doesn‚Äôt make sense to create Account outside a Bank

````python
class Bank:

    def __init__(self,bank_name, ifsc_code, address):
        self.bank_name = bank_name
        self.ifsc_code = ifsc_code
        self.address = address

    def bank_details(self):
        print(f"Bank: {self.bank_name}")
        print(f"IFSC Code: {self.ifsc_code}")
        print(f"Address: {self.address}")

    # Inner Class

    class Account:

        def __init__(self, holder_name, balance):
            self.holder_name = holder_name
            self.balance = balance

        def account_details(self, bank_obj):
            bank_obj.bank_details()
            print(f"Account Holder Name: {self.holder_name}")
            print(f"Balance: {self.balance}")

        def deposite_amount(self,amount):
            self.balance += amount
            print(f"Balance: {self.balance}")
````

````python
# Creating Bank object
b1 = Bank("ICICI Bank", "ICIC0000007", "Connaught Place Branch")
b1.bank_details()
````

````python
# Creating Inner Class object
acc1 = b1.Account("Anushaka", 145000)
acc1.account_details(b1) # Inner class can access outer class instance using reference.
````

````python
acc1.deposite_amount(15000)
````
