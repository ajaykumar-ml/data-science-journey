# **Three Fundamental OOP Terms Class, Object, and Reference Variable**

#### What is Class ?
A class is a blueprint or template that defines what data (attributes) and behavior (methods) an object will have.

#### Definition of Object
An object is a real instance of a class that occupies memory and contains actual data and behavior defined by the class.

#### Definition of Reference Variable
A reference variable is a variable that refers to an object by storing the memory address (reference) of that object.

## Understanding Class, Object, and Reference Variable  
### (Society Planning Real-Life Analogy)

To understand Object Oriented Programming concepts clearly,  
we can relate them to a **real-life society planning scenario**.

---

### üîπ Class ‚Äî Society Plan (Blueprint)

A **class** is like the **plan or blueprint of a society**.

- The society plan defines:
  - Number of buildings
  - Layout of roads
  - Parks and common facilities
- The plan itself:
  - Does not occupy land
  - Has no residents
  - Is only a design

üëâ In OOP terms:
> **Society Plan = Class**

The class defines **what an object will look like and what it can do**,  
but it does not represent a real entity by itself.

---

### üîπ Object ‚Äî Constructed Society

An **object** is the **actual society constructed using the plan**.

- Using the same society plan, we can construct societies:
  - In Delhi
  - In Bangalore
  - In Jaipur
- Each constructed society:
  - Occupies real land
  - Has real buildings and residents
  - Is independent from other societies

üëâ In OOP terms:
> **Each constructed society = Object**

Even though the plan (class) is the same,  
each object has its **own existence and data**.

---

### üîπ Reference Variable ‚Äî Society Name

A **reference variable** is the **name by which we refer to a society**.

- Examples:
  - "Green Valley Society"
  - "Sunrise Enclave"

These names:
- Are not the society itself
- Only point to the actual society

üëâ In Python:
> **Reference variable stores the reference (address) of an object**

Multiple reference variables can also point to the **same object**.

---

### üîÅ Mapping Real Life to OOP

| Real Life Concept | OOP Concept |
|------------------|------------|
| Society plan | Class |
| Constructed society | Object |
| Society name | Reference variable |

---

### üß† Key Takeaway

- A **class** is a blueprint or plan
- An **object** is a real instance created from the class
- A **reference variable** points to the object in memory

Just like one society plan can be used to build multiple societies in different cities,  
one class can be used to create multiple objects in a program.
