# What is a Nested Dictionary?
1. A nested dictionary is a dictionary that contains another dictionary as its value.
2. Nested dictionaries are used when data has structure and hierarchy.

```python
student = {
    "name": "ajay",
    "academics": {
        "database":89,
        "operating_system":82,
        "computer_network":90,
        "machine_learning":99,
        "deep_learning":99,
        "linear_algebra":100
    }
}
student
```

### Accessing Data from Nested Dictionary

**outer level**

```python
student["name"]
```

**inner level**

```python
student["academics"]["deep_learning"]
```

### Updating Nested Dictionary Values

```python
student["academics"]["operating_system"] = 98
print(student["academics"]["operating_system"])
```

```python
student["academics"]["data_structure"] = 83
student
```

### Looping Through Nested Dictionaries in Python
1. Real-life meaning
    - Outer key → Student ID
    - Inner dictionary → Student details

```python
record = {
    101: {"name":"ajay", "marks":92, "city":"delhi"},
    102: {"name":"anushka", "marks":90, "city":"pune"},
    103: {"name":"sunny", "marks":85, "city":"jaipur"},
    104: {"name":"sonali","marks":78,"city":"pune"}
}

print("Outer loop:")
for students_ids in record:
    print(students_ids)
```

```python
print("Inner loop:")
for students_ids, details in record.items():
    print(students_ids, details)
```

**Deep Nested Dictionary (Level-2 Example)**
1. **Task 1:** Count Employees per Department
2. **Task 2:** Find High-Salary Employees (> 60,000)
3. **Task 3:** Filter Experienced Employees (≥ 5 years)

```python
company = {
    "IT": {
        "E101": {"name": "Ajay", "salary": 60000, "experience": 5},
        "E102": {"name": "Ravi", "salary": 55000, "experience": 4},
        "E103": {"name": "Neha", "salary": 65000, "experience": 6},
        "E104": {"name": "Amit", "salary": 50000, "experience": 3}
    },

    "HR": {
        "E201": {"name": "Pooja", "salary": 48000, "experience": 4},
        "E202": {"name": "Karan", "salary": 52000, "experience": 5},
        "E203": {"name": "Sneha", "salary": 47000, "experience": 3}
    },

    "Finance": {
        "E301": {"name": "Rahul", "salary": 70000, "experience": 7},
        "E302": {"name": "Anita", "salary": 68000, "experience": 6}
    },

    "Sales": {
        "E401": {"name": "Vikas", "salary": 45000, "experience": 2},
        "E402": {"name": "Nisha", "salary": 50000, "experience": 3},
        "E403": {"name": "Arjun", "salary": 52000, "experience": 4}
    }
}
```

**Task 1: Count Employees per Department**

```python
for dept,employees in company.items():
    print(dept, "has", len(employees), "employees")
```

**Task 2: Find High-Salary Employees (> 60,000)**

```python
for dept,employees in company.items():
    for ids, details in employees.items():
        if details["salary"] >= 60000:
            print(details["name"], details["salary"], dept)
```

**Task 3:Filter Experienced Employees (≥ 5 years)**

```python
for dept , employees in company.items():
    for ids, details in employees.items():
        if details["experience"] >= 5:
            print(details["name"], details["experience"],details["salary"], dept)
```
