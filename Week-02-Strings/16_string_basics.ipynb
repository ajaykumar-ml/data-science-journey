### Slicing in Strings
Slicing means extracting a part of string.

string [begin : end : step]<br>
1. begin → included
2. end → excluded
3. step → jump
4. Note:<br>
a. If we are not specifying begin index then it will consider from beginning of the string.<br>
b. If we are not specifying end index then it will consider up to end of the string.<br>
c. The default value for step is 1.

```python
s = "Python"

print(s[0:4])   # Pyth
print(s[2:6])   # thon
print(s[:3])    # Pyt
print(s[3:])    # hon
```

## Behaviour of Slice Operator:
1) string [begin : end : step]
2) Step value can be either +ve or –ve
3) If +ve then it should be forward direction(left to right) and we have to consider begin
to end-1
4) If -ve then it should be backward direction (right to left) and we have to consider begin
to end+1.

**Note:**
1. In the backward direction if end value is -1 then result is always empty.
2. In the forward direction if end value is 0 then result is always empty.

**Note**
1. Slice operator never raises **IndexError**

**Slice Operator Case Study:**

```python
text = "Data Science is the field that combines mathematics, programming, and domain knowledge"
```

```python
print(text[1:6:2]) # beg = 1, end-1 = 5, step = 2 , forward direction

print(text[::1]) # beg = 0, end-1 = 85, step = 1 , forward direction

print(text[::-1]) # step = -ve means backward direction, reverse of given string

print(text[::0]) # step cant not be zero
```
