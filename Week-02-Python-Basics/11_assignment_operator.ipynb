### Assignment operators
1. Assignment operators are used to assign values to variables.
2. Python also provides compound assignment operators that combine an operation with assignment.
3. Compound Assignment Operators
| Operator | Meaning               | Example   | Same As      |
| -------- | --------------------- | --------- | ------------ |
| `+=`     | Add and assign        | `x += 5`  | `x = x + 5`  |
| `-=`     | Subtract and assign   | `x -= 3`  | `x = x - 3`  |
| `*=`     | Multiply and assign   | `x *= 2`  | `x = x * 2`  |
| `/=`     | Divide and assign     | `x /= 4`  | `x = x / 4`  |
| `//=`    | Floor divide & assign | `x //= 2` | `x = x // 2` |
| `%=`     | Modulus & assign      | `x %= 3`  | `x = x % 3`  |
| `**=`    | Power & assign        | `x **= 2` | `x = x ** 2` |
| `&=`     | and & assign          | `x &= 2`  | `x = x and 2`|



```python
x = 10
x += 5
print(x)
```

```python
x = 20
x -= 3
print(x)
```

```python
x = 30
x *= 2
print(x)
```

```python
x = 40
x /= 4
print(x)
```

```python
x = 55
x //= 2
print(x)
```

```python
x = 65
x %= 2
print(x)
```

```python
x = 2
x **= 10
print(x)
```

```python
x = 5
x &= 2
print(x)
```

### Multiple assignment

```python
x, y, z = 1, 2, 3
print(x,y,z)
```

### Assignment with Swapping (Python Special)

```python
a = 5
b = 10

print("a = ",a, "b = ", b)
a, b = b, a
print("After Swapping")
print("a = ",a, "b = ", b)
```

###  Why x++/x-- is not allowed in Python?
1. Python does not support ++ and -- because integers are immutable.
2. Python avoids pre/post increment operators to keep syntax simple and readable. Instead, += 1 is used.

### What about ++x and --x in Python?
1. +x â†’ unary plus (does nothing)
2. +(+x) â†’ still the same value
3. -(-x) â†’ Double negation â†’ original value.

```python
x = 10
y = 20
print(++x)
print(-y)
print(--y)
```

### Difference Between Unary, Binary, and Ternary Operators in Python
1.
| Operator Type | Operands | Example            |
| ------------- | -------- | ------------------ |
| Unary         | 1        | `-x`, `not x`      |
| Binary        | 2        | `a + b`            |
| Ternary       | 3        | `a if cond else b` |

2.
. Unary operator works on one operand <br>
. Binary operator works on two operands <br>
. Ternary operator works on three operands using condition

3. Ternary Operator (3 operands)<br>
. Uses three operands<br>
. Used for conditional expression<br>
. Syntax: (value_if_true if condition else value_if_false)


```python
x = 10
y = 20

min = x if (x<y) else y

print(min)
```
