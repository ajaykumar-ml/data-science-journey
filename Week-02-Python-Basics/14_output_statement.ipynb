### print() is used to display output on the screen in Python.
| Feature         | Example       |
| --------------- | ------------- |
| Basic           | `print("Hi")` |
| Multiple values | `print(a, b)` |
| Separator       | `sep=","`     |
| Line end        | `end=" "`     |
| Formatting      | `f"{x}"`      |


```python
a = 10
b = 20
print(a) # Print  a value ,  Moves cursor to next line
print(b)
```

```python
a = 30
b = 20
print(a,b) # By default, values are separated by space
print(a,b, sep=":")
```

```python
print("ajay")
print("kumar")

print("Use Line end")

print("ajay", end=" ")
print("kumar")
```

```python
a , b,  c, d = 10,20,30,40
print(a,b,c,d,sep="::")
```

```python
print(a,b,c,d,sep="&&", end="^^")
print(d,c,b,a, sep="###")
# By default end is new line
print(a,b,c,d,sep="&&")
print(d,c,b,a, sep="###")
```

### Formatted Output (VERY IMPORTANT)


a) **f-string (Best & Modern)**

```python
eid = 1
name  = "ajay"
age  = 25

print(f"Name: {name}, Age:{age}")
```

```python
eno = int(input("Enter Employee no :"))
name = input("Enter Employee name :")
age = int(input("Enter Employee  age:"))
city = input("Enter Employee  City:")
mrrd = eval(input("Enter Employee  married [True/False]:"))

print("Please confirm Employee Details")
print(f"ID: {eno}\n Name: {name}\n Age: {age}\n City:{city}\n Married:{mrrd}")
```

b) **format() method**

```python
name = "ajay"
age = 25
print("Name: {}, Age: {}".format(name,age))
```

```python
eno = int(input("Enter Employee no :"))
name = input("Enter Employee name :")
age = int(input("Enter Employee  age:"))
city = input("Enter Employee  City:")
mrrd = eval(input("Enter Employee  married [True/False]:/n"))

print("Please confirm Employee Details")
print("ID={}\n Name={}\n Age={}\n City={}\n Married={}".format(eno,name,age,city,mrrd))

# Using Index
# print("ID={0}\n Name={1}\n Age={2}\n City={3}\n Married={4}".format(eno,name,age,city,mrrd))

# Using Kewords
# print("ID={x}\n Name={y}\n Age={z}\n City={a}\n Married={b}".format(x = eno,y = name,z = age,a = city,b =mrrd))
```
