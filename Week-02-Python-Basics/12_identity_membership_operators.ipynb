### Identity Operators in Python (is, is not)

1. Identity operators are used to check whether two variables point to the same object in memory, not whether their values are equal.

```python
a = 10
b = 10
print(a is b)
print(a is not b)
```

```python
a = "ajay"
b = "ajay"
print(a is b)
print(a is not b)
```

```python
a = 10 + 20j
b = 10 + 20j
print(a is not b)
```

### ⚠️ Identity vs Equality (VERY IMPORTANT)
1. Equality (==) → value comparison
2. Identity (is) → memory comparison

```python
l1 = [10,20,30]
l2 = [10,20,30]
print(l1 is l2) # address comparison
print(id(l1),id(l2)) #addresses are different
print(l1 == l2) # content comparison
```

# Membership Operators in Python (in, not in)

1. Membership operators are used to check whether a value exists inside a sequence or collection.
2. Works With These Data Types
| Data Type  | Example                       |
| ---------- | ----------------------------- |
| List       | `[1, 2, 3]`                   |
| Tuple      | `(1, 2, 3)`                   |
| String     | `"python"`                    |
| Set        | `{1, 2, 3}`                   |
| Dictionary | `dict` (checks **keys only**) |

3. Performance Note (Extra Knowledge) <br>
. List / Tuple / String → Linear search (slow for large data) <br>
. Set / Dict → Hash-based (very fast)


```python
nums = [1, 2, 3, 4]

print(2 in nums)     
print(10 in nums)    
```

```python
text = "python"

print("p" in text)     
print("z" in text)  
print("P" in text)
```

```python
nums = [1, 2, 3]

print(5 not in nums)   
print(2 not in nums)   
```

```python
student = {"name": "Ajay", "surname": "Kumar", "age": 25}

print("name" in student)   
print("Ajay" in student)   
print("Kumar" in student)
print("surname" in student)
```
