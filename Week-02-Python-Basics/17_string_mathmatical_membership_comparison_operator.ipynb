### Mathematical opertaors

1. Strings support only + (concatenation) and * (repetition). Other mathematical operators are not allowed.<br>

**Note:**
1) To use + operator for Strings, compulsory both arguments should be str type.<br>
2) To use * operator for Strings, compulsory one argument should be str and other
argument should be int.

```python
s1 = "Data"
s2 = "Science"

result = s1 + s2
print(result)

result = s1 + " " + s2 # with space
print(result)
```

### Membership Operators on Strings
1. Membership operators check presence of characters or substrings.
2. Operators Used <br>
a). in <br>
b). not in
3. Membership checks are case-sensitive.

```python
text = "Data Science is the field that combines mathematics, programming, and domain knowledge"
print("field" in text)
print("maths" not in text)
print("mathematics" in text)
print("Mathematics" in text)
```

### Comparison Operators on Strings
1. Comparison operators compare strings lexicographically (i.e., based on ASCII / Unicode values).
2. 

| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | Equal            |
| `!=`     | Not equal        |
| `<`      | Less than        |
| `>`      | Greater than     |
| `<=`     | Less or equal    |
| `>=`     | Greater or equal |

**What is ord()?**<br>
1. The ord() function returns the Unicode (ASCII) integer value of a single character.
2. The ord() function helps understand:<br>
    . Why sorting of strings behaves differently for uppercase and lowercase<br>
    . Text preprocessing logic

***Convert a string into Unicode values:***

```python
text = "Data Science"
for x in text:
    print("Unicode value of {} is: {}".format(x,ord(x)))
```

```python
text = "aAbBcCdD123"
for x in text:
    print("Unicode of {} is: {}".format(x, ord(x)))
```
